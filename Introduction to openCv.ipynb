# image rotation and flipping

```python
# syntax : M = cv2.getRotationMatrix2D(center, angle, scale)
#           rotated_image = cv2.wrapAffine(image, M, (width, height))
(h,w) = color_img.shape[:2]
center = (h//2, w//2)
M = cv2.getRotationMatrix2D(center, 45, 2.0)
rotated_img = cv2.warpAffine(color_img, M, (w, h))
cv2.imshow("origional_img",color_img)
cv2.imshow("rotated_image", rotated_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# flipping : cv2.flip(image, fliped) fliped means =0(vertically),1(horoizontally) or -1(both)
flipped_img = cv2.flip(color_img, -1) 
cv2.imshow("origional_img",color_img)
cv2.imshow("flipped_image", flipped_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Filter from Scratch

```python
 shape = cv2.imread(r"shape.jpg")

cv2.imshow("shape", shape)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
shape.shape
```

```python
shape_bw = cv2.cvtColor(shape,cv2.COLOR_BGR2GRAY)
print(shape.shape,shape_bw.shape, sep="\n")
```

```python
cv2.imshow("black white shapes", shape_bw)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
vertical_fltr = np.tile(np.array([-1,0,1]),(3,1))
print(vertical_fltr)
```

```python
horizontal_fltr = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
print(horizontal_fltr)
```

# for vertical shape 

```python
h = shape_bw.shape[0]
w = shape_bw.shape[1]

i=0
j=0

shape_vertical_fltr = []

while i<h-2:
    t = []
    j=0
    while j<w-2:
        t.append(np.multiply(shape_bw[i:i+3,j:j+3],vertical_fltr).sum())
        j+=1
    shape_vertical_fltr.append(t)
    i+=1

shape_vertical = np.array(shape_vertical_fltr)


shape_vertical.shape
```

```python
plt.imshow(shape_vertical)
```

```python
shape_vertical_final = shape_vertical.astype(np.uint8)
cv2.imshow("black white shapes", shape_vertical_final)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# for horizontal shape

```python
h = shape_bw.shape[0]
w = shape_bw.shape[1]

i=0
j=0

shape_horizontal_fltr = []

while i<h-2:
    t = []
    j=0
    while j<w-2:
        t.append(np.multiply(shape_bw[i:i+3,j:j+3],horizontal_fltr).sum())
        j+=1
    shape_horizontal_fltr.append(t)
    i+=1

shape_horizontal = np.array(shape_horizontal_fltr)


shape_horizontal.shape
```

```python
plt.imshow(shape_horizontal)
```

```python
shape_horizontal_final = shape_horizontal.astype(np.uint8)
cv2.imshow("black white shapes", shape_horizontal_final)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Applying both filter together

```python
vertical_result = cv2.filter2D(shape_bw,-1,vertical_fltr)
horizontal_result = cv2.filter2D(shape_bw,-1,horizontal_fltr)

# combine
combined_result = np.abs(vertical_result) + np.abs(horizontal_result)

cv2.imshow("Combined Result",combined_result)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Video Processing 
### video componant
1. video - it is the combination of images
2. webcam - A webcam is a digital camera that captures video and audio data and transmits it in real-time over the internet.
3. frame - 1 photo is the part of video chain and 1 video is multiple frames
4. frame by frame pricessing - Frame-by-frame processing in computer vision is the technique of treating a video as a series of individual images (frames) and applying image processing techniques to each frame sequentially.

### upload any video using cv2.VideoCapture(source)

```python
import cv2
```

```python
cap = cv2.VideoCapture(0)
while True:
    ret, frame = cap.read()  #ret = true/false, frame = image
    if not ret:
        print("could not read frame")
        break

    cv2.imshow("webcam feed", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        print("Quitting....")
        break

cap.release()
cv2.destroyAllWindows()
```
